```python
from urllib.request import urlopen
url = "http://bpms.kemco.or.kr/transport_2012/car/car_choice.aspx?serorigin=&serecode=&serfuel=&sergear=&sercartype=&sergrade=&sertypegb=&sertab=&otype=&ptype=200&f=system&sermileage1=&sermileage2=&serco21=&serco22=&seramt1=&seramt2=&sermname=&serstyear=#search_top"
htmlData = urlopen(url).read()
```

```python
type(htmlData)
```

```python
htmlData = htmlData.decode()
type(htmlData)
```

```python
htmlData
```

HTML 파싱


```python
from bs4 import BeautifulSoup as bs
html = bs(htmlData, 'html.parser')
html
```

### 테이블 내용 추출


```python
tr = html.select('#search_top > table > tbody > tr')
tr
```

### 파일저장

```python
with open("car.csv", 'w') as f:
  line = '업체명,모델명,표시연비,배기량,CO2배출량,등급,연간유류비\n'
  f.write(line)
  for item in tr:
    tds = item.find_all('td')
    tds3 = tds[3].text.replace('.','')
    tds6 = tds[6].text.replace(',','')
    level = tds[5].select_one('img')['src']
    gidx = level.index('.gif') - 1
    line = f'{tds[0].text.strip()},{tds[1].text},{float(tds[2].text)},{tds3},{int(tds[4].text)},{int(level[gidx])},{int(tds6)}'
    print(line)
    f.write(line)
```
